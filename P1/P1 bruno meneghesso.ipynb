# Primeiro Conceito

## Paises que entram na união européia tendem a ter melhores indicadores economicos

### Indices avaliados:

GDP total

GDP capita

GDP trabalhador

exportações 

importações

### Momentos avalidaos:

anos logo antes da formação da união 1992 (formação dos 3 pilares)

anos logo apos à formação da união européia 

atualmete

### Grupos de paises avaliados:

EU (grupo teste)

paises europeus fora da EU (grupo de controle)

europa como um todo (indicador de tendencia)

### Hipótese de método


As parcerias econômicas gerads pela EU proporcionam maior desenvolvimento dos paises participantes.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gdp = pd.read_excel('gdp_total_yearly_growth.xlsx')
gdp = gdp.set_index('country')
gdpcap = pd.read_excel('gdppercapita_us_inflation_adjusted.xlsx')
gdpcap = gdpcap.set_index('country')
gdptrab = pd.read_excel('gdpperemployee_us_inflation_adjusted.xlsx')
gdptrab = gdptrab.set_index('country')
exp = pd.read_excel('exports_percent_of_gdp.xlsx')
exp = exp.set_index('country')
imp = pd.read_excel('imports_percent_of_gdp.xlsx')
imp = imp.set_index('country')
```

```python
europa = ['Russia','Germany','United Kingdom','France','Italy','Spain','Ukraine','Poland','Romania','Netherlands','Belgium','Greece','Czechia','Portugal','Sweden','Hungary','Belarus','Austria','Serbia','Switzerland','Bulgaria','Denmark','Finland','Slovakia','Norway','Ireland','Croatia','Moldova','Bosnia and Herzegovina','Albania','Lituania','North Macedonia','Slovenia','Latvia','Estonia','Montenegro','Luxemburg','Malta','Iceland','Andorra','Monaco','Liechtenstein','San Marino','Holy See']
gdp_euro = gdp.loc[europa]
gdpcap_euro = gdpcap.loc[europa]
gdptrab_euro = gdptrab.loc[europa]
exp_euro = exp.loc[europa]
imp_euro = imp.loc [europa]
```

```python
EU = ['Belgium','Greece','Lithuania','Portugal','Bulgaria','Spain','Luxenbourg','Romania','Czech Republic','France','Hungary','Denmark','Croatia','Malta','Slovakia','Germany','Italy','Netherlands','Finland','Estonia','Cyprus','Austria','Sweden','Ireland','Latvia','Poland','United Kingdom']
noEU = []
for e in europa:
    if e not in EU:
        noEU.append(e)
gdp_EU = gdp.loc[EU]
gdpcap_EU = gdpcap.loc[EU]
gdptrab_EU = gdptrab.loc[EU]
exp_EU = exp.loc[EU]
imp_EU = imp.loc [EU]
gdp_noEU = gdp.loc[noEU]
gdpcap_noEU = gdpcap.loc[noEU]
gdptrab_noEU = gdptrab.loc[noEU]
exp_noEU = exp.loc[noEU]
imp_noEU = imp.loc [noEU]
```

```python
#EUROPA
gdp_euro_pre = gdp_euro[[1989,1990,1991,1992]]
gdpcap_euro_pre = gdpcap_euro[[1989,1990,1991,1992]]
gdptrab_euro_pre = gdptrab_euro[[1991,1992]] #dados não existentes para datas anteriores
exp_euro_pre = exp_euro[[1989,1990,1991,1992]]
imp_euro_pre = imp_euro[[1989,1990,1991,1992]]

gdp_euro_pos = gdp_euro[[1992,1993,1994,1995]]
gdpcap_euro_pos = gdpcap_euro[[1992,1993,1994,1995]]
gdptrab_euro_pos = gdptrab_euro[[1992,1993,1994,1995]] 
exp_euro_pos = exp_euro[[1992,1993,1994,1995]]
imp_euro_pos = imp_euro[[1992,1993,1994,1995]]

#4 anos mais recentes disponiveis
gdp_euro_now = gdp_euro[[2010,2011,2012,2013]]
gdpcap_euro_now = gdpcap_euro[[2014,2015,2016,2017]]
gdptrab_euro_now = gdptrab_euro[[2014,2015,2016,2017]] 
exp_euro_now = exp_euro[[2014,2015,2016,2017]]
imp_euro_now = imp_euro[[2014,2015,2016,2017]]
```

```python
#EU
gdp_EU_pre = gdp_EU[[1989,1990,1991,1992]]
gdpcap_EU_pre = gdpcap_EU[[1989,1990,1991,1992]]
gdptrab_EU_pre = gdptrab_EU[[1991,1992]] #dados não existentes para datas anteriores
exp_EU_pre = exp_EU[[1989,1990,1991,1992]]
imp_EU_pre = imp_EU[[1989,1990,1991,1992]]

gdp_EU_pos = gdp_EU[[1992,1993,1994,1995]]
gdpcap_EU_pos = gdpcap_EU[[1992,1993,1994,1995]]
gdptrab_EU_pos = gdptrab_EU[[1992,1993,1994,1995]] 
exp_EU_pos = exp_EU[[1992,1993,1994,1995]]
imp_EU_pos = imp_EU[[1992,1993,1994,1995]]

#4 anos mais recentes disponiveis
gdp_EU_now = gdp_EU[[2010,2011,2012,2013]]
gdpcap_EU_now = gdpcap_EU[[2014,2015,2016,2017]]
gdptrab_EU_now = gdptrab_EU[[2014,2015,2016,2017]] 
exp_EU_now = exp_EU[[2014,2015,2016,2017]]
imp_EU_now = imp_EU[[2014,2015,2016,2017]]
```
